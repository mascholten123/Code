## Example curves for Methodology section of Report 

```python

import numpy as np
import matplotlib.pyplot as plt

# Generating sample data for pump curve
flow_rate_pump = np.arange(0, 0.1, 0.01)  # Flow rate in cubic meters per second (m³/s)
head_pump = 100 - 8000 * flow_rate_pump ** 2  # Quadratic pump curve in meters (m)

# Generating sample data for system curve
head_system = 40 + 6000 * flow_rate_pump ** 2  # Quadratic system curve in meters (m)

# Plotting pump curve and system curve
plt.plot(flow_rate_pump, head_pump, label='Pump Curve', color='blue')
plt.plot(flow_rate_pump, head_system, label='System Curve', color='orange')

# Adding labels and title
plt.title('Pump Curve and System Curve')
plt.xlabel('Flow Rate (m³/s)')
plt.ylabel('Head (m)')
plt.legend()

# Displaying the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Define the pump curve and system curve functions
def pump_curve(flow_rate):
    return 100 - 8000 * flow_rate ** 2

def system_curve(flow_rate):
    return 40 + 6000 * flow_rate ** 2

# Find the intersection point
def intersection(flow_rate):
    return pump_curve(flow_rate) - system_curve(flow_rate)

operation_point = fsolve(intersection, 0.05)[0]

# Generating sample data for pump curve
flow_rate_pump = np.arange(0, 0.1, 0.01)  # Flow rate in cubic meters per second (m³/s)
head_pump = pump_curve(flow_rate_pump)

# Generating sample data for system curve
head_system = system_curve(flow_rate_pump)

# Plotting pump curve and system curve
plt.plot(flow_rate_pump, head_pump, label='Pump Curve', color='blue')
plt.plot(flow_rate_pump, head_system, label='System Curve', color='orange')

# Marking the intersection point
plt.plot(operation_point, pump_curve(operation_point), 'ko')
plt.text(operation_point - 0.003, pump_curve(operation_point), 'Operation Point', ha='right')

# Adding labels and title
#plt.title('Pump Curve and System Curve')
plt.xlabel('Flow Rate (m³/s)')
plt.ylabel('Head (m)')
plt.legend()

# Displaying the plot
plt.grid(True)
plt.show()
```
