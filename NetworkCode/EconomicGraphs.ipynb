# Comparing Scenarios

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

size = np.array([1, 2, 3])
NPC = np.array([174634.05, 242032.76, 331897.70])
CI = np.array([144573.16, 203365.10, 281754.34])

plt.plot(size, NPC, marker='^', color='tab:blue')
plt.plot(size, CI, marker='^', color='tab:orange')
plt.plot(1, 238942.84 , marker='x', color='tab:blue')
plt.plot(1, 123526.22 , marker='x', color='tab:orange')

# Custom legend handles and labels
custom_lines = [Line2D([0], [0], marker='o', color='w', markerfacecolor='tab:blue', markersize=10),
                Line2D([0], [0], marker = 'o', color='w', markerfacecolor='tab:orange', markersize=10),
                Line2D([0], [0], marker='^', color='tab:blue' , markersize=10),
                Line2D([0], [0], marker='x', color='tab:blue', markersize=10)]

plt.legend(custom_lines, ['NPC', 'Initial Investment Cost', 'Direct Drive System', 'Battery System'], loc='upper center', bbox_to_anchor=(0.45, 1.05), ncol=2, fancybox=True, shadow=True)

plt.ylabel("[DOP]")
plt.xlabel("Plot Size [ha]")
plt.xticks([1, 2, 3])
plt.show()

```
